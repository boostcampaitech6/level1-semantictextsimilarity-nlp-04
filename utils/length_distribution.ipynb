```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/train.csv')

# 'label'이 0.0인 행만 선택
df_zero_label = df[df['label'] == 0.0].copy()

# 'sentence_1'과 'sentence_2'의 길이를 계산
df_zero_label.loc[:, 'sentence_1_len'] = df_zero_label['sentence_1'].apply(len)
df_zero_label.loc[:, 'sentence_2_len'] = df_zero_label['sentence_2'].apply(len)
```

```python
# 시각화
plt.figure(figsize=(12, 6))

plt.suptitle('label == 0.0 : Length Distribution')

plt.subplot(1, 2, 1)
sns.histplot(df_zero_label['sentence_1_len'], kde=False, bins=30)
plt.title('sentence_1 Length Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df_zero_label['sentence_2_len'], kde=False, bins=30)
plt.title('sentence_2 Length Distribution')

plt.tight_layout()
plt.show()
```
